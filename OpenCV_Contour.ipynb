```python
import cv2

img = cv2.imread("triangle.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

# Find Contours 

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Display Output
cv2.drawContours(img, contours, -1, (0,255,0), 3)

cv2.imshow("Contours", img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Detect Shapes

```python
import cv2

img = cv2.imread("triangle.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

# Find Contours 

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
     # Approximate shape
    perimeter = cv2.arcLength(cnt, True)
    approx = cv2.approxPolyDP(cnt, 0.02 * perimeter, True)
    corners = len(approx)

    # Determine shape
    if corners == 3:
        shape = "Triangle"
    elif corners == 4:
        shape = "Rectangle"
    elif corners == 5:
        shape = "Pentagon"
    elif corners >5:
        shape = "Circle"
    else:
        shape = "Unknown"

# Display Output
    cv2.drawContours(img, [approx], -1, (0,255,200), 3)

    x= approx.ravel()[0]            # Ravel converts nD array in 1D array
    y = approx.ravel()[1]-10

    cv2.putText(img, shape, (x,y), cv2.FONT_HERSHEY_COMPLEX,0.8, (0,0,200), 2)

cv2.imshow("Contours", img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Circle

```python
import cv2

img = cv2.imread("circle.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

_, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)

# Find Contours 

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

for cnt in contours:
     # Approximate shape
    perimeter = cv2.arcLength(cnt, True)
    approx = cv2.approxPolyDP(cnt, 0.01 * perimeter, True)
    corners = len(approx)

    # Determine shape
    if corners == 3:
        shape = "Triangle"
    elif corners == 4:
        shape = "Rectangle"
    elif corners == 5:
        shape = "Pentagon"
    elif corners >5:
        shape = "Circle"
    else:
        shape = "Unknown"

# Display Output
cv2.drawContours(img, [approx], -1, (0,255,200), 2)

x= approx.ravel()[0]            # Ravel converts nD array in 1D array
y = approx.ravel()[1]-10

cv2.putText(img, shape, (x,y), cv2.FONT_HERSHEY_COMPLEX,0.8, (0,0,255), 2)

cv2.imshow("Contours", img)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
