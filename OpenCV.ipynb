### Convert to Gray-Scale

```python
img = cv2.imread(r"C:\Users\Jivan.Patil\Downloads\images.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow("Gray", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Convert between color spaces

```python
# Convert between color spaces
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # BGR to RGB
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # BGR to Grayscale
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)  # BGR to HSV

# Display different color spaces
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

axes[0,0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axes[0,0].set_title('Original (BGR)')

axes[0,1].imshow(img_gray, cmap='gray')
axes[0,1].set_title('Grayscale')

axes[1,0].imshow(img_rgb)
axes[1,0].set_title('RGB')

axes[1,1].imshow(img_hsv)
axes[1,1].set_title('HSV')

plt.tight_layout()
plt.show()
```

## Fonts

```python
# Create image to demonstrate fonts
img = np.zeros((500, 800, 3), dtype=np.uint8)

# List of available fonts
fonts = [
    cv2.FONT_HERSHEY_SIMPLEX,
    cv2.FONT_HERSHEY_PLAIN,
    cv2.FONT_HERSHEY_DUPLEX,
    cv2.FONT_HERSHEY_COMPLEX,
    cv2.FONT_HERSHEY_TRIPLEX,
    cv2.FONT_HERSHEY_COMPLEX_SMALL,
    cv2.FONT_HERSHEY_SCRIPT_SIMPLEX,
    cv2.FONT_HERSHEY_SCRIPT_COMPLEX
]

font_names = [
    "FONT_HERSHEY_SIMPLEX",
    "FONT_HERSHEY_PLAIN", 
    "FONT_HERSHEY_DUPLEX",
    "FONT_HERSHEY_COMPLEX",
    "FONT_HERSHEY_TRIPLEX",
    "FONT_HERSHEY_COMPLEX_SMALL",
    "FONT_HERSHEY_SCRIPT_SIMPLEX",
    "FONT_HERSHEY_SCRIPT_COMPLEX"
]

# Display each font
y_position = 40
for font, name in zip(fonts, font_names):
    cv2.putText(img, name, (20, y_position), font, 0.7, (255, 255, 255), 2)
    y_position += 50

plt.figure(figsize=(12, 8))
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Available OpenCV Fonts')
plt.axis('off')
plt.show()
```

## fontScale- Font Size

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

img = np.zeros((400, 600, 3), dtype=np.uint8)

# Different font scales
scales = [0.5, 1.0, 1.5, 2.0, 3.0, 4.0]
y_position = 50

for scale in scales:
    text = f'Scale: {scale}'
    cv2.putText(img, text, (20, y_position), 
                cv2.FONT_HERSHEY_SIMPLEX, scale, (255, 255, 255), 2)
    y_position += int(60 * scale)  # Adjust spacing based on font size

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Different Font Scales')
plt.show()
```

## lineType - Line Type (Optional)

```python
img = np.zeros((300, 600, 3), dtype=np.uint8)

# Different line types
line_types = [cv2.LINE_4, cv2.LINE_8, cv2.LINE_AA]
line_type_names = ["LINE_4", "LINE_8", "LINE_AA (Anti-aliased)"]

y_position = 50
for line_type, name in zip(line_types, line_type_names):
    cv2.putText(img, name, (20, y_position), 
                cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 0), 2, line_type)
    y_position += 80

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Different Line Types')
plt.show()
```

```python
import cv2

# Read an image
img = cv2.imread(r"C:\Users\Jivan.Patil\Downloads\images.jpg")

# Get image dimensions
height, width = img.shape[:2]

# Define rotation center
center = (width // 2, height // 2)

# Create rotation matrix (angle=65, scale=1.0)
rotation_matrix = cv2.getRotationMatrix2D(center, 65, 1)

# Apply rotation
rotated_65 = cv2.warpAffine(img, rotation_matrix, (height, height))

# Display images
# cv2.imshow('Original Image', img)
cv2.imshow('Rotated 65 Degrees', rotated_65)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
