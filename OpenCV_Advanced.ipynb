## Draw a rectangle

```python
import cv2
import numpy as np

flag = False
ix = -1
iy = -1

def draw(event, x, y, flags, param):
    global flag, ix, iy, img_copy

    if event == 1:
        flag = True
        ix = x
        iy = y
    
    elif event == 0:
        if flag == True:
            img_copy = img.copy()
            cv2.rectangle(img_copy, (ix, iy), (x,y), color=(0,0,255), thickness=2)
            cv2.imshow("Smart Window", img_copy)

    elif event ==4:
        flag = False
        cv2.rectangle(img, (ix, iy), (x,y), color=(0,0,255), thickness=2)
        cv2.imshow("Smart Window", img)


cv2.namedWindow("Smart Window")

cv2.setMouseCallback("Smart Window", draw)

img = np.zeros((512,512,3))

cv2.imshow("Smart Window", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

drawing = False   # True when mouse button is pressed
ix, iy = -1, -1   # Starting point

# Mouse callback function
def draw_rectangle(event, x, y, flags, param):
    global ix, iy, drawing, img, img_copy

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix, iy = x, y

    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            img_copy = img.copy()
            cv2.rectangle(img_copy, (ix, iy), (x, y), (0, 0, 255), 2)
            cv2.imshow("Smart Window", img_copy)

    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img, (ix, iy), (x, y), (0, 0, 255), 2)
        cv2.imshow("Smart Window", img)

# Create window
cv2.namedWindow("Smart Window")

# # Create black image
img = np.zeros((512, 512, 3), dtype=np.uint8)
#img_copy = img.copy()

# Set mouse callback
cv2.setMouseCallback("Smart Window", draw_rectangle)

cv2.imshow("Smart Window", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Working with Videos using OpenCV

```python
import cv2

capture = cv2.VideoCapture(0)



while True:
    retrn, frame = capture.read() #retrn = True/False, frame= Image
    cv2.imshow("webcam", frame)
    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cv2.destroyAllWindows()
```

### Convert to B/W

```python
import cv2

capture = cv2.VideoCapture(0)



while True:
    retrn, frame = capture.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    cv2.imshow("webcam", gray)

    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

cv2.destroyAllWindows()
```

### Save/ Record webcam video

```python


import cv2

capture = cv2.VideoCapture(0)
fourcc = cv2.VideoWriter_fourcc(*'MJPG')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640, 480))

while True:
    retrn, frame = capture.read()
    out.write(frame)
    cv2.imshow("webcam", frame)

    if cv2.waitKey(1) & 0xFF == ord('x'):
        break

out.release()
cv2.destroyAllWindows()
```
