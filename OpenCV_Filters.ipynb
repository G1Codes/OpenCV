# Gaussian Blur


```python
import cv2

path = "images.jpg"
img = cv2.imread(path)

gaussian_blur = cv2.GaussianBlur(img, (7,7), 0)

cv2.imshow("blurred", gaussian_blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Median Blur

```python
import cv2

path = "images.jpg"
img = cv2.imread(path)

med_blur = cv2.medianBlur(img, 5)

cv2.imshow("blurred", med_blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Sharpening

```python
import cv2
import numpy as np

path = "low_res.jpg"

img = cv2.imread(path)


sharp_kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])

sharp = cv2.filter2D(img, -1, sharp_kernel)

cv2.imshow("Original", img)
cv2.imshow("Sharpened", sharp)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Even stronger sharpening:

```python
import cv2
import numpy as np

path = "low_res.jpg"

img = cv2.imread(path)


sharp_kernel = np.array([[ -1, -1, -1],
                   [ -1,  9, -1],
                   [ -1, -1, -1]])

sharp = cv2.filter2D(img, -1, sharp_kernel)

cv2.imshow("Original", img)
cv2.imshow("Sharpened", sharp)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
