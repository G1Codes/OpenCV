# ‚ö° BITWISE OPERATIONS - "IMAGE MATH"

### Draw a circle and rectangle on 2 images:

```python
import cv2
import numpy as np

img1 = np.zeros((300,300), dtype= "uint8")
img2 = np.zeros((300,300), dtype= "uint8")

#Draw a circle on image 1:
cv2.circle(img1, (150,150), 100, color= 255, thickness=-1)

#Draw a rectangle on image 2:
cv2.rectangle(img2, (100,100), (250,250), 255, -1)

cv2.imshow("Image1", img1)
cv2.imshow("Image2", img2)

cv2.waitKey(0)
```

### Apply bitwise operations

```python
import cv2
import numpy as np

img1 = np.zeros((300,300), dtype= "uint8")
img2 = np.zeros((300,300), dtype= "uint8")

#Draw a circle on image 1:
cv2.circle(img1, (150,150), 100, color= 255, thickness=-1)

#Draw a rectangle on image 2:
cv2.rectangle(img2, (100,100), (250,250), 255, -1)

bitwise_and = cv2.bitwise_and(img1, img2)
bitwise_or = cv2.bitwise_or(img1, img2)
bitwise_xor = cv2.bitwise_xor(img1, img2)
bitwise_not = cv2.bitwise_not(img1)


cv2.imshow("bitwise_and", bitwise_and)
cv2.imshow("bitwise_or", bitwise_or)
cv2.imshow("bitwise_xor", bitwise_xor)
cv2.imshow("bitwise_not", bitwise_not)

cv2.waitKey(0)
```
